* What is a Cost Function?
A Cost Function measures how wrong the model's predictions are.
It tells us:
"How much error is there between the actual value and the predicted value?"
Error = Actual - Predicted
Error = 90 - 80 = 10

* What is Gradient Descent?
Gradient Descent is an optimization algorithm used to reduce the Cost Function and find the best model (best-fit line).
Simple meaning:
It helps the model “learn” by reducing errors step by step.

(Easy Intuition)

Imagine you are on top of a mountain and want to go down to the lowest point (valley).

Height of mountain = Cost (error)
Bottom of valley = Minimum error (best model)

You don’t know the path, so you:

 take small steps downward
 check direction
 keep going until you reach the lowest point

That process = Gradient Descent

* R² Score tells how well your regression model is performing.
It measures:
“How much of the variation in the output is explained by the model?”
How good is my model compared to a “dumb model”?
R² = 1   → Perfect model (best)
R² = 0   → Worst (same as average prediction)
R² < 0   → Very bad model

*Bias is an error when the model is too poor it dosnot trained well with data. it occurs high bias low variance.

* variance when the model is trained too good but the new data poor in result. high variance.

Good Model = Balance between Bias and Variance, giving accurate predictions on both training and new data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Model complexity (x-axis)
complexity = np.arange(1, 11)

# Simulated curves
bias = 100 / complexity          # Bias decreases as complexity increases
variance = complexity ** 2       # Variance increases as complexity increases
total_error = bias + variance    # Total error curve

# Plot
plt.figure(figsize=(8,5))

plt.plot(complexity, bias, label="Bias (High → Low)", linewidth=2)
plt.plot(complexity, variance, label="Variance (Low → High)", linewidth=2)
plt.plot(complexity, total_error, label="Total Error", linewidth=3)

plt.title("Bias vs Variance Tradeoff (Python Visualization)")
plt.xlabel("Model Complexity (Simple → Complex)")
plt.ylabel("Error")
plt.legend()
plt.grid(True)

plt.show()
```

4) Polynomial Regression?
Polynomial Regression is used when the relationship between input and output is NOT a straight line.
Solution: Polynomial Regression
Polynomial Regression fits a curve instead of a straight line.
Example
House Price vs House Age
House Age	Price
1	         300k
5	         280k
10	         250k
20	         180k
The relationship may be curved, not linear.
So we use Polynomial Regression.

Why Use Polynomial Regression?
When:
Scatter Plot shows a curve
instead of a straight line.
Then Linear Regression gives poor predictions.
Polynomial Regression can capture the curve.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# Sample data
X = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([2, 5, 10, 17, 26, 37])

# Linear Regression
linear_model = LinearRegression()
linear_model.fit(X, y)

# Polynomial Regression (degree=2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

poly_model = LinearRegression()
poly_model.fit(X_poly, y)

# Smooth curve for plotting
X_new = np.linspace(1, 6, 100).reshape(-1, 1)
X_new_poly = poly.transform(X_new)

# Plot
plt.scatter(X, y, label="Actual Data")
plt.plot(X_new, linear_model.predict(X_new), label="Linear Regression")
plt.plot(X_new, poly_model.predict(X_new_poly), label="Polynomial Regression")

plt.title("Linear vs Polynomial Regression")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
# Gini Impurity measures how mixed (impure) a node is, not counting purity and impurity separately.

import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0, 1, 100)

gini = 1 - (p**2 + (1-p)**2)

plt.plot(p, gini)

plt.title("Gini Impurity")
plt.xlabel("Probability of Yes")
plt.ylabel("Gini")

plt.grid(True)
plt.show()
```

6) Cross Validation is a technique used to evaluate a machine learning model more reliably.
Instead of testing the model only once, we test it multiple times using different splits of the data.

A single Train-Test split may give misleading results because performance depends on how the data was split.
Cross Validation reduces this problem.

Definition

The dataset is divided into:

Training Set → Used to train the model
Testing Set → Used to evaluate the model

100 Records

80 Records → Training
20 Records → Testing

Cross Validation repeatedly splits the data and evaluates the model multiple times.

Benefits

More reliable accuracy

Better model evaluation

Uses data efficiently

Helps detect overfitting

What is Stratified K-Fold?
Stratified K-Fold preserves the original class distribution in every fold and is commonly used for classification problems with imbalanced data.
